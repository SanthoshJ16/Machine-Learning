```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Reading the Dataset
dataset = pd.read_csv('insurance_pre.csv')
dataset=pd.get_dummies(dataset,drop_first=True)

# Creating input variables and output
independent=dataset[['age','bmi', 'children','sex_male', 'smoker_yes']]
dependent=dataset[['charges']]

#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size = 0.3, random_state = 0)
```

```python

from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(max_depth=3, random_state=42)
from sklearn.ensemble import AdaBoostRegressor
best_score = -np.inf
regressor = AdaBoostRegressor(
    estimator=regressor,     # for scikit-learn >= 1.2; use base_estimator on older versions
    n_estimators=10,        # number of weak learners
    learning_rate=0.1,       # contribution of each learner
    random_state=42,
    loss="exponential"            # options: "square", "linear", "exponential"
)
regressor.fit(X_train, np.ravel(y_train))
y_pred=regressor.predict(X_test)
from sklearn.metrics import r2_score
r_score=r2_score(np.ravel(y_test),y_pred)
r_score
```

```python
import pickle
filename="finalModel_AdaRegressor.sav"
pickle.dump(regressor,open(filename,"wb"))
```

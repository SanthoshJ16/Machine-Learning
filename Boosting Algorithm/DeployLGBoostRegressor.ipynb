```python
#Deployment model for the sales prediction based on Advertising spend
import pickle

#load the final predicted Sales model and store in variable load_model
load_model=pickle.load(open("finalModel_LGBoostRegressor.sav",'rb'))


# Ask user for input and store it on variable
age_input=float(input("Enter you Age: "))
bmi_input=float(input("Enter your BMI: "))
children_input=float(input("Enter Number of Children: "))
sex_input=int(input("Enter your Sex 0 or 1:"))
smoker_input=int(input("Do you Smoke ? 0 or 1:"))

# Run prediction using dynamic input 
result=load_model.predict([[age_input,bmi_input,children_input,sex_input,smoker_input]])

print("Predicted Insurance Charges is:", result)
```

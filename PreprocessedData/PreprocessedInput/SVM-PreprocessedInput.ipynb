```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Reading the Dataset
dataset = pd.read_csv('50_Startups.csv')
```

```python
#dataset
```

```python
datasets=pd.get_dummies(dataset,drop_first=True)
```

```python
#datasets
```

```python
independent=datasets[['R&D Spend','Administration', 'Marketing Spend','State_Florida', 'State_New York']]
dependent=datasets[['Profit']]
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size =0.30, random_state = 0)
```

```python
#standardizing the input training and test set
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

```

```python
from sklearn.svm import SVR
regressor=SVR(kernel="rbf",C=1000)
regressor=regressor.fit(X_train,y_train.values.ravel())
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

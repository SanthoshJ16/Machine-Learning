```python
#Deployment model for the  50 Startup Profit prediction based on Preprocessed Inputs
import pickle
import pandas as pd

# Reading the Dataset
dataset = pd.read_csv('50_Startups.csv')
datasets=pd.get_dummies(dataset,drop_first=True)

independent=datasets[['R&D Spend','Administration', 'Marketing Spend','State_Florida', 'State_New York']]
dependent=datasets[['Profit']]

#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size =0.30, random_state = 0)

#standardizing the input training and test set
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train.values)
X_test = sc.transform(X_test.values)

## Preprocessing the output

#standardizing the output training and test set
from sklearn.preprocessing import StandardScaler
scy = StandardScaler()
y_train = scy.fit_transform(y_train)
y_test = scy.transform(y_test)

## Get the Input dynamically
rd_input=float(input("R&D:"))
admin_input=float(input("Admin"))
mark_input=float(input("Marketing "))
fol_input=int(input("State_flo 0 or 1:"))
new_input=int(input("State_new 0 or 1:"))

# Preprocessing the input data
#preinput=sc.transform([[1300,12000,4000,0,1]])
preinput=sc.transform([[rd_input,admin_input,mark_input,fol_input,new_input]])
#preinput

```

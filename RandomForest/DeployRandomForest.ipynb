```python
#Deployment model for the sales prediction based on Advertising spend
import pickle

#load the final predicted Sales model and store in variable load_model
load_model=pickle.load(open("finalModel_RandomForest.sav",'rb'))


# Ask user for input and store it on variable
rd_input=float(input("R&D:"))
admin_input=float(input("Admin"))
mark_input=float(input("Marketing "))
fol_input=int(input("State_flo 0 or 1:"))
new_input=int(input("State_new 0 or 1:"))

# Run prediction using dynamic input 
result=load_model.predict([[rd_input,admin_input,mark_input,fol_input,new_input]])

print("Predicted Profit is:", result)
```

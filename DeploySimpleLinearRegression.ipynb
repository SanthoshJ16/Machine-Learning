```python
#Deployment model for the sales prediction based on Advertising spend
import pickle

#load the final predicted Sales model and store in variable load_model
load_model=pickle.load(open("final_sales_prediction_model.sav",'rb'))


# Ask user for input and store it on variable value
value = float(input("Enter the value for Advertising spend: "))

# Run prediction using dynamic input to return the Sales Prediction and store in variable result
result = load_model.predict([[value]])

print("Sales Prediction for Advertising spend amount is:", result)
```

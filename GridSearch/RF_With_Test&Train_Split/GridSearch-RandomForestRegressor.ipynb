```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Reading the Dataset
dataset = pd.read_csv('insurance_pre.csv')
dataset=pd.get_dummies(dataset,drop_first=True)

# Creating input variables and output
independent=dataset[['age','bmi', 'children','sex_male', 'smoker_yes']]
dependent=dataset[['charges']]

#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size = 1/3, random_state = 0)


#standardizing the input training and test set
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)


#Import the GridSearchCV and load the params (Hyper Tuning Parameters)
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
param_grid = {'criterion':['squared_error', 'absolute_error', 'friedman_mse', 'poisson'],
'max_features': ['sqrt','log2',None],
'n_estimators':[10,50]}


grid = GridSearchCV(RandomForestRegressor(), param_grid, refit = True, verbose= 3,n_jobs=-1)
# fitting the model for grid search
grid.fit(X_train, y_train.values.ravel())

grid_predictions = grid.predict(X_test)

# print r2 Score
from sklearn.metrics import r2_score
r_score=r2_score(y_test.values.ravel(),grid_predictions)
print(r_score)

# Print the best model and the Hypertuning Parameter
re=grid.cv_results_
print("The R_score value for best parameter {}:".format(grid.best_params_))

```

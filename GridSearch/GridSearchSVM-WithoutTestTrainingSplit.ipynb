```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Reading the Dataset
dataset = pd.read_csv('insurance_pre.csv')
dataset=pd.get_dummies(dataset,drop_first=True)

# Creating input variables and output
independent=dataset[['age','bmi', 'children','sex_male', 'smoker_yes']]
dependent=dataset[['charges']]

#split into training set and test - Removed Code 
#from sklearn.model_selection import train_test_split
#X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size = 1/3, random_state = 0)


#standardizing the input training and test set
#from sklearn.preprocessing import StandardScaler
#sc = StandardScaler()
#X_train = sc.fit_transform(X_train)
#X_test = sc.transform(X_test)


#Import the GridSearchCV and load the params (Hyper Tuning Parameters)
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid = {'kernel':['rbf','poly','sigmoid','linear'],
'C':[10,100,1000,2000,3000],'gamma':['auto','scale']}
grid = GridSearchCV(SVR(), param_grid, refit = True, verbose = 3,n_jobs=-1)
# fitting the model for grid search
grid.fit(independent, dependent) ## Directly passing the Inputs and Outputs since GRID evalutes all the combination
# based on Hypertunig parameters, no need to split test and train

# Print the best rscore value for the best Hypertuning Parameter
re=grid.cv_results_
print("The R_score value for best parameter {}:".format(grid.best_params_))

```

```python

##Put the results in a table and print the table 
table=pd.DataFrame.from_dict(re)
#table
```
